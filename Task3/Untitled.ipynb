```python
import matplotlib.pyplot as plt

def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    steps = []  # To store the search steps

    while left <= right:
        mid = (left + right) // 2
        steps.append((left, right, mid))

        if arr[mid] == target:
            return mid, steps  # Element found
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1, steps  # Element not found

def plot_binary_search(arr, target):
    result, steps = binary_search(arr, target)

    if result != -1:
        print(f"Element {target} found at index {result}.")
    else:
        print(f"Element {target} not found in the list.")

    # Create a plot to visualize the binary search steps
    plt.figure(figsize=(12, 6))
    plt.title("Binary Search")
    plt.xlabel("Array Index")
    plt.ylabel("Array Value")

    # Plot the array
    plt.plot(range(len(arr)), arr, label="Array")

    # Highlight the search steps
    for left, right, mid in steps:
        plt.axvline(x=mid, color="red", linestyle="--")
        plt.axvline(x=left, color="gray", linestyle="--")
        plt.axvline(x=right, color="gray", linestyle="--")

    plt.legend()
    plt.show()

# Example usage:
my_list = [1, 3, 4, 5, 6, 8, 9, 11, 15, 17, 19]
target_element = 8
plot_binary_search(my_list, target_element)
```

```python
def bubble_sort(arr):
    n = len(arr)

    # Traverse through all elements in the list
    for i in range(n):
        # Flag to optimize the algorithm by detecting if any swaps occurred in this pass
        swapped = False

        # Last i elements are already in place, so we don't need to compare them
        for j in range(0, n - i - 1):
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # Swap the elements
                swapped = True  # Set the flag to True as a swap occurred

        # If no two elements were swapped in the inner loop, the array is already sorted
        if not swapped:
            break

# Example usage:
my_list = [64, 34, 25, 12, 22, 11, 90]
print("Original list:", my_list)

bubble_sort(my_list)
print("Sorted list:", my_list)
```

```python
def quick_sort(arr):
    if len(arr)<=1:
        return arr
    else:
        pivot = arr[len(arr)//2]
        left = [x for x in arr if x<pivot]
        mid = [x for x in arr if x==pivot]
        right = [x for x in arr if x>pivot]

        return quick_sort(left) + mid + quick_sort(right)

my_list = [3, 6, 8, 10, 1, 2, 1]
sorted_list = quick_sort(my_list)
print("Original list:", my_list)
print("Sorted list:", sorted_list)
```
