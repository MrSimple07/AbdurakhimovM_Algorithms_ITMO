```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx  # for graph visualisation
```

```python
graph = nx.generators.random_graphs.gnm_random_graph(n=100, m=200)

plt.figure(figsize=(10, 10))
plt.title('Graph Visualization')
nx.draw(graph, with_labels=True, node_color='tab:blue')
```

```python
adjacency_matrix = nx.adjacency_matrix(graph).todense()
adjacency_matrix
```

```python
dict_of_lists = nx.to_dict_of_lists(graph)

for key in dict_of_lists.keys():
    print('{}: {}'.format(key, dict_of_lists[key]))
    
    if key == 9:
        break
```

```python
def depth_first(graph, source):
    return nx.dfs_preorder_nodes(
        G=graph,
        source=source
    )

def breadth_first(graph, source, target):
    return nx.bidirectional_shortest_path(
        G=graph,
        source=source,
        target=target
    )
```

```python
depth_first_args = list(depth_first(graph, 0))

plt.figure(figsize=(10, 10))
plt.title('Graph Visualization')
nx.draw_shell(graph.subgraph(depth_first_args), with_labels=True, node_color='tab:blue')

print(depth_first_args)
```

```python
breadth_first_args = list(breadth_first(graph, 0, len(graph)-1))

plt.figure(figsize=(10, 10))
plt.title('Graph Visualization')
nx.draw_shell(graph.subgraph(breadth_first_args), with_labels=True, node_color='tab:blue')

print(breadth_first_args)
```
